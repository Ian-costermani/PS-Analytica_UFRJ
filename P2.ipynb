```python
def calculadora_angulo():
    while True:
        print("Digite uma hora no formato - 'hh:mm'\n")
        tempo = input()
        #caso queira que o programa pare
        if tempo == 'f':
            print("Fim...")
            break
        #verificando se a entrada é valida
        try:
            if len(tempo) != 5 or tempo[2] != ':':
                raise ValueError
            if not tempo[0].isdigit() or not tempo[1].isdigit() or not tempo[3].isdigit() or not tempo[4].isdigit():
                raise ValueError

            #retirando a hora e os minutos da entrada
            horas, minutos = map(int, tempo.split(':'))

            if horas >= 24 or horas < 0 or minutos >= 60 or minutos < 0:
                print("Input Inválido\n")
                continue
            #transformando a hora
            if horas >= 12:
                horas -= 12
            angulo_hora = (360 / 12) * horas  
            angulo_minuto = (360 / 60) * minutos 

            #calculando o menor angulo
            angulo = abs(angulo_hora - angulo_minuto)

            menor_angulo = min(angulo, 360 - angulo)

            print(f"O menor ângulo é de {menor_angulo:.0f}°\n")
        
        except ValueError:
            print("Input Inválido\n")

            
```

```python
calculadora_angulo()       
```

```python
def movimentos_cavalo():
    
    
    linhas = [1, 2, 3, 4, 5, 6, 7, 8]
    colunas = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h']
    traducao = {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5, 'f': 6, 'g': 7, 'h': 8}
    movimentos = [
        (2, 1), (2, -1), (-2, 1), (-2, -1),
        (1, 2), (1, -2), (-1, 2), (-1, -2)
    ]

    while True:
        print("Digite uma posição no tabuleiro de xadrez (ex: d4)\n")
        posicao = input()
        # caso queira que o programa pare
        if posicao == 'f':
            print("Fim...")
            break

        try:
            #verificando a entrada
            if len(posicao) != 2 or posicao[0] not in colunas or int(posicao[1]) not in linhas:
                raise ValueError

            coluna_inicial = traducao[posicao[0]]
            linha_inicial = int(posicao[1])

            # calculando os possíveis movimentos do cavalo
            possiveis_posicoes = []
            for mov in movimentos:
                nova_coluna = coluna_inicial + mov[0]
                nova_linha = linha_inicial + mov[1]
                
                # Verificando se a nova posição é possivel
                if 1 <= nova_coluna <= 8 and 1 <= nova_linha <= 8:
                    # convertendo o número da coluna de volta para uma letra
                    posicao_nova = colunas[nova_coluna - 1] + str(nova_linha)  
                    possiveis_posicoes.append(posicao_nova)

            # Exibindo as posições possíveis separadas por vírgulas
            print(", ".join(possiveis_posicoes))

        except ValueError:
            print("Input inválido\n")

movimentos_cavalo()
```

```python
def calcular_notas_moedas():
    # definindo as notas e moedas que vao ser trabalhadas
    notas = [100, 50, 20, 10, 5, 2]
    moedas = [1.00, 0.50, 0.25, 0.10, 0.05, 0.01]

    while True:
        print("Digite um valor com apenas duas casas decimais (ex: 123.73):")
        valor_input = input()
        if valor_input == 'f':
            print("Fim...")
            break
        
        try: 
            # verificando se a entrada é valida
            if len(valor_input) < 4 or valor_input[-3] != '.':
                print("Input inválido, deve haver 2 casas decimais.")
                continue
            
            # convertendo a entrada para um float 
            valor = float(valor_input)
            if valor < 0:
                raise ValueError
            
            # calculando as notas
            print("NOTAS:")
            for nota in notas:
                num_notas = valor // nota     # calcula o número de notas
                valor -= num_notas * nota  #atualiza o valor restante
                print(f"{int(num_notas)} nota(s) de R$ {nota:.2f}")

            # calculando as moedas
            print("MOEDAS:")
            for moeda in moedas:
                num_moedas = valor // moeda #calcula o número de moedas
                valor -= num_moedas * moeda # atualiza o valor restante
                print(f"{int(num_moedas)} moeda(s) de R$ {moeda:.2f}")

        except ValueError:
            print("Input inválido\n")


calcular_notas_moedas()
```

```python
def frequencia_numeros():
    #dicionario para guardar os valores
    numeros = {}  
    while True:
        entrada = input("Digite um número inteiro ou 'f' para finalizar:\n")
        # caso o programa pare
        if entrada == 'f':
            ordem = sorted(numeros)  
            for i in ordem:
                print(f"O número {i} apareceu {numeros[i]} vezes")  
            print("Fim...")
            return 
        # ignorando entradas não válidas
        if not entrada.isdigit():
            continue
        # convertendo a entrada para um inteiro
        n = int(entrada)
        
        # atualizando o dicionário de frequência
        if n not in numeros:
            numeros[n] = 1
        else:
            numeros[n] += 1

frequencia_numeros()
```
